# Data Cleaning - Pokémon Dataset

Before starting any analysis, the dataset needs to be inspected and cleaned to ensure consistency and reliability.

This step focuses on identifying missing values, removing unnecessary columns, and standardizing the dataset so it can be used for further analysis and modeling.

```python
import pandas as pd

df = pd.read_csv("../data/raw/pokemon.csv")
df.head()
```

```python
df.info()
df.isnull().sum()
```

## Initial Data Inspection

The dataset was first loaded and explored to understand its structure, data types, and potential issues.

Key checks performed:
- Overview of columns and data types
- Identification of missing values
- Detection of unnecessary or inconsistent columns

At this stage, the goal was not to modify the data, but to observe and understand it.

```python
df = df.drop(columns=["Unnamed: 0"])
```

## Removing Irrelevant Columns

The dataset contained an extra column named `Unnamed: 0`, which appears to be an index column from a previous export.

Since it does not provide any meaningful information for analysis, it was removed to keep the dataset clean and focused.

```python
df["Type 2"] = df["Type 2"].fillna("None")
```

## Handling Missing Values

The column `Type 2` contains missing values. This is expected because not all Pokémon have a secondary type.

Instead of removing these rows or leaving them as null, the missing values were replaced with `"None"` to clearly indicate the absence of a second type.

This helps maintain consistency and avoids issues during analysis and modeling.

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

## Checking for Duplicate Records

The dataset was checked for duplicate rows to ensure data integrity.

If duplicates were found, they were removed to prevent skewed analysis. In this case, the dataset either contained no duplicates or they were handled accordingly.

```python
df.columns = df.columns.str.strip().str.replace(" ", "_")
```

## Standardizing Column Names

Column names were cleaned by:
- Removing leading/trailing spaces
- Replacing spaces with underscores

This makes the column names easier to work with in code and avoids potential errors when referencing them.

```python
df.to_csv("../data/cleaned/pokemon_cleaned.csv", index=False)
```

## Saving the Cleaned Dataset

After cleaning, the dataset was saved as a new file in the `data/cleaned/` directory.

This ensures that:
- The original raw data remains unchanged
- All further analysis is performed on a consistent and prepared dataset
